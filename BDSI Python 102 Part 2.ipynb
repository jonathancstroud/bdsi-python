## Prettier plots

Right now, our plot is pretty uninformative. Which line is which? What do the axes mean? We can add more information to our plots easily with Matplotlib.

### axis labels and titles

````python
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('My title')
plt.xlabel('My x-axis')
plt.ylabel('My y-axis')
````

### Legends

````python
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['x**2', 'x**3', 'x**4'])
````

You can embed LaTeX directly into the plots.

````python
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['$x^2$', '$x^3$', '$x^4$'])
````

You can move the legend around as needed

````python
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['$x^2$', '$x^3$', '$x^4$'], loc='lower right')
````

### Changing the figure size/aspect ratio

````python
plt.figure(figsize=(16,9))
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['$x^2$', '$x^3$', '$x^4$'])
plt.show()

plt.figure(figsize=(3,6))
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['$x^2$', '$x^3$', '$x^4$'])
````

### Changing the axes limits

````python
plt.figure(figsize=(10,10))
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['$x^2$', '$x^3$', '$x^4$'])
plt.xlim([-1,1])
plt.ylim([-1,1])
````

### Colors, Markers, and Sizes

`plot` and `scatter` each take an additional string argument that allows you to set the color and style properties of the line or points that you draw.

This argument takes the form `'<color><markerstyle>'`, and there are many different options for each.

#### Common Colors

 - `b` : blue
 - `g` : green
 - `r` : red
 - `c` : cyan
 - `m` : magenta
 - `y` : yellow
 - `k` : black
 - `w` : white
 
#### Line Styles (for `plot`)

 - `-` : line
 - `--` : dashed line
 - `-.` : dash/dotted line
 - `:` : dotted line

#### Marker Styles

 - `.` : point
 - `o` : circle
 - `s` : square
 - `p` : pentagon
 - `d` : diamond
 - `*` : star
 - `+` : plus
 - `x` : x

Here are some examples:

````python
plt.figure(figsize=(10,10))
plt.plot(x, y0, 'r-')  # red line
plt.plot(x, y1, 'b--') # blue dashed line
plt.plot(x, y2, 'k:')  # black dotted line
plt.legend(['$x^2$', '$x^3$', '$x^4$'])
````

#### 1. Produce a plot of the functions:

 - $f(x) = e^{x/10}\sin(\pi x)$
 - $g(x) = xe^{x/3}$
 
 over the interval [0, 10]. Include a legend.

````python
import math # You'll need math.exp, math.pi, math.sin, math.cos
````

````python
# Start by defining a list of x-values from 0 to 10
xs = [x/10.0 for x in range(0,100)]
````

````python
f = [math.exp(-x/10.0)*math.sin(math.pi*x) for x in xs]
g = [x*math.exp(-x/3.0) for x in xs]
plt.plot(xs, f)
plt.plot(xs, g)
plt.legend(['$f(x)$', '$g(x)$'])
````

### 2. Plotting a limacon

The shape of a limacon can be defined parametrically as:

$x = (r_0 + \cos \theta) \cos \theta$

$y = (r_0 + \cos \theta) \sin \theta$

When $r_0 = 1$, this curve is called a cardioid. Use this definition to plot
the shape of a limacon for $r_0 = 0.8$, $r_0 = 1.0$, and $r_0 = 1.2$. Be sure to use enough points that the curve is closed and appears smooth (except for
the cusp in the cardioid).

````python
thetas = [theta/100.0 for theta in range(0, 628)]
````

````python
plt.figure(figsize=(10,10))

x = [(0.8 + math.cos(theta)) * math.cos(theta) for theta in thetas]
y = [(0.8 + math.cos(theta)) * math.sin(theta) for theta in thetas]
plt.plot(x, y)

x = [(1.0 + math.cos(theta)) * math.cos(theta) for theta in thetas]
y = [(1.0 + math.cos(theta)) * math.sin(theta) for theta in thetas]
plt.plot(x, y)

x = [(1.2 + math.cos(theta)) * math.cos(theta) for theta in thetas]
y = [(1.2 + math.cos(theta)) * math.sin(theta) for theta in thetas]
plt.plot(x, y)
plt.legend(['$r_0 = 0.8$', '$r_0 = 1.0$', '$r_0 = 1.2$'])
````
